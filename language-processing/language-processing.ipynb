```python
#17
def get_last(text_list, word, index, prev_index) -> int:
    if index == 0:
        prev_index = index
        index = text_list.index(word, index)
    else:
        prev_index = index
        try:
            index = text_list.index(word, index+1)
        except ValueError:
            return prev_index
    return get_last(text_list, word, index, prev_index)

text9_list = [w for w in text9]
last_idx = text9_list.index('.', text9_list.index('sunset'))
first_idx = text9_list.index('.', get_last(text9_list[:643], '.',0,None),text9_list.index('sunset'))
text9_list[first_idx+1:last_idx+1]
```
